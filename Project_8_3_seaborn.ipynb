### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn offers a wide range of plot types to handle various data visualization tasks effectively.

Commonly used plots that we can create using Seaborn and their uses:

1. Scatter Plot:
Scatter plots are used to visualize the relationship between two continuous variables. They help identify patterns, trends, and correlations between the variables. Seaborn's scatter plot can also incorporate additional information, such as color-coding points based on a third categorical variable, making it more informative.

2. Bar Plot:
Bar plots are effective for visualizing and comparing categorical data. They are used to display the distribution or count of categories and to compare the values of different categories. Seaborn's bar plot can also display confidence intervals, making it useful for presenting statistical summaries.

3. Box Plot (Box-and-Whisker Plot):
Box plots provide a graphical summary of the distribution of numerical data. They show the median, quartiles, and the presence of outliers in the data. Seaborn's box plot is helpful for identifying data spread, skewness, and potential outliers across multiple groups or categories.

4. Heatmap:
Heatmaps are used to display a 2D representation of data where individual values are represented by colors. They are useful for visualizing the correlation matrix, showing patterns in large datasets, and identifying clusters or trends in the data. Seaborn's heatmap is especially popular for displaying correlation matrices.

5. Violin Plot:
Violin plots combine aspects of a box plot and a kernel density plot. They provide a better representation of the data's distribution, including information about the density of values. Seaborn's violin plot is useful for comparing the distribution of data between different categories.


### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
fmri_data = sns.load_dataset("fmri")
```

```python
fmri_data.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="timepoint" , y = "signal" ,data = fmri_data,hue="event",style="region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot")
plt.show()
```

### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data.head(3)
```

```python
#pclass vs age
sns.boxplot(x="pclass",y="age",data = titanic_data)
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot: pclass vs. age')


#pclass vs. fare
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot: pclass vs. fare')

plt.show()
```

### Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flight_data = sns.load_dataset("flights")
```

```python
flight_data.head()
```

```python
flights_matrix = flight_data.pivot(index='month', columns='year', values='passengers')
sns.heatmap(flights_matrix,cmap='coolwarm', annot=True, fmt='d')
plt.title('Flights Heatmap')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
